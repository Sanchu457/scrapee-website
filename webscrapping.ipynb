```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
with open('Amazon Scraping - Sheet1.csv', 'r') as data:
    csv_reader = csv.DictReader(data, delimiter=',')
    for lines in csv_reader:
        url_string = 'https://www.amazon.' + lines['country'] + '/dp/' + lines['Asin']
        response=requests.get(url_string)       
        if response.status_code ==404:
            print(url_string + ' not available')
        if response.status_code ==503:
                print(url_string + ' is invalid')
        soup = BeautifulSoup(response.content, "html.parser")
        res = soup.find_all('span',attrs={'class':'ProductTitle'})
        res1 = soup.find_all('span',attrs={'class':'ProductImageUrl'})
        res2 = soup.find_all('span',attrs={'class':'PriceoftheProduct'})
        res3 = soup.find_all('span',attrs={'class':'ProductDetails'})
        type(res)



```
